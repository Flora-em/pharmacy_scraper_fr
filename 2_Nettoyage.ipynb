## ETAPE 1 : Assemblage et homogénéisation 

### Fusion fichier Flora

```python
# Ajout de la variable "region" pour mes fichiers
# Pour coïncider avec les autres fichiers
dossier = "avis_regions"

fichiers_et_regions = {
    "avis_auvergne_rhone_alpes_FINAL.csv": "Auvergne-Rhone-Alpes",
    "avis_bourgogne_franche_comte.csv": "Bourgogne-Franche-Comte",
    "avis_centre_val_de_loire.csv": "Centre-Val de Loire",
    "avis_bretagne.csv": "Bretagne",
    "avis_corse.csv": "Corse",}

for nom_fichier, nom_region in fichiers_et_regions.items():
    chemin_fichier = os.path.join(dossier, nom_fichier)
    if os.path.exists(chemin_fichier):
        df = pd.read_csv(chemin_fichier, sep=";", encoding="utf-8-sig")
        df["region"] = nom_region
        df.to_csv(chemin_fichier, sep=";", index=False, encoding="utf-8-sig")
        print(f"{chemin_fichier} : colonne 'region' ajoutee ({nom_region}), {len(df)} lignes")
    else:
        print(f"{chemin_fichier} : introuvable, ignore")
```

```python
dossier = "avis_regions"

fichiers_a_fusionner_Flo = [
    "avis_auvergne_rhone_alpes_FINAL.csv",
    "avis_bourgogne_franche_comte.csv",
    "avis_centre_val_de_loire.csv",
    "avis_bretagne.csv",
    "avis_corse.csv"
]
```

```python
morceaux_Flo = []
for fichier in fichiers_a_fusionner_Flo:
    chemin_fichier = os.path.join(dossier, fichier)
    if os.path.exists(chemin_fichier):
        df = pd.read_csv(chemin_fichier, sep=";", encoding="utf-8-sig")
        morceaux_Flo.append(df)
        print(f"{chemin_fichier} : {len(df)} lignes")
    else:
        print(f"{chemin_fichier} : introuvable, ignore")

df_regions_Flo = pd.concat(morceaux_Flo, ignore_index=True)
print("\nTotal fusion :", len(df_regions_Flo), "lignes")
```

```python
chemin_sortie = os.path.join(dossier, "avis_regions_Flo.csv")
df_regions_Flo.to_csv(chemin_sortie, sep=";", index=False, encoding="utf-8-sig")
print("Fichier sauvegarde :", chemin_sortie)
```
